```python
total_pop = ky['전체 인구수']
senior_pop = ky_melted[ky_melted['구분'] == '65세이상'].groupby('경기도시군')['인구수'].sum()
ratio = (senior_pop / total_pop).fillna(0)
```

```python
plt.figure(figsize=(10, 6))
ax = sns.histplot(
    data=ky_melted,
    x='경기도시군',
    hue='구분',
    weights='인구수',  # ky_melted에 존재하는 컬럼
    multiple="dodge",
    shrink=0.8)

plt.title('경기도 인구 분포')
plt.xlabel('경기도시군')
plt.ylabel('인구수(백만)')
plt.xticks(rotation=45)
plt.tight_layout()

xticks = ax.get_xticks()
xticklabels = [tick.get_text() for tick in ax.get_xticklabels()]

for i, city in enumerate(xticklabels):
    if city in ratio:        
        percent = ratio[city] * 100       
        heights = [p.get_height() for p in ax.patches if p.get_x() <= xticks[i] <= p.get_x() + p.get_width()]
        if heights:
            y = max(heights)
            ax.text(
                xticks[i], y + y * 0.01, f'{percent:.1f}%', 
                ha='center', va='bottom', fontsize=9, color='red')
plt.show()
```
